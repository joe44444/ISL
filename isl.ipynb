```python
#DATAVIZ
#Importing packages
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Joe Rishwanth\Desktop\new\Assignment Files\ada.csv')
```

```python
#SHOT CONVERSION RATE
df1 = df[['player_id','player_name','tour_id','goals','shots']].copy()
```

```python
df2=df1.iloc[0:2674]
```

```python
df3=df1.iloc[2674:4860]
df3
```

```python
gr=df2.groupby(['player_name'])['goals','shots'].sum()
```

```python
a=gr.sort_values(by='goals', ascending=False).head()
```

```python
a['shot conversion rate'] = a['shots']/a['goals']
```

```python
al=a.reset_index()
al
```

```python
#LEAGUE1
# Setting the positions and width for the bars
pos = list(range(len(al['goals']))) 
width = 0.25 
# Plotting the bars
fig, ax = plt.subplots(figsize=(10,5))
plt.bar(pos, 
        al['goals'], 
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        label=al['player_name'][0]) 


plt.bar([p + width for p in pos], 
        al['shots'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        label=al['player_name'][1]) 

plt.bar([p + width*2 for p in pos], 
       al['shot conversion rate'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        label=al['player_name'][2]) 

# Set the y axis label
ax.set_ylabel('')

# Set the chart's title
ax.set_title('Shot conversion rate of top 5 goal scorers')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(al['player_name'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(al['goals'] + al['shots'] + al['shot conversion rate'])] )

# Adding the legend and showing the plot
plt.legend(['goals', 'shots', 'shot conversion rate'], loc='upper left')
plt.grid()
plt.show()
```
